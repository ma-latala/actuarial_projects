```python
import yfinance as yf
```

```python
tech_tickers = [
    "AAPL",   # Apple
    "MSFT",   # Microsoft
    "GOOGL",  # Alphabet (Google)
    "AMZN",   # Amazon — not pure tech but behaves like one in markets
    "META",   # Meta Platforms (Facebook)
    "NVDA",   # Nvidia
    "TSLA",   # Tesla — tech-ish, high-beta, interesting chart patterns
    "AMD",    # Advanced Micro Devices
    "INTC",   # Intel
    "CRM"     # Salesforce
]
```

```python
# Download historical data for the S&P 500 index

import numpy as np

ticker = "^GSPC"
start_date = "2015-12-07"
end_date   = "2025-12-05"

data = yf.download(ticker, start=start_date, end=end_date)

prices = data["Close"]

log_returns = np.log(prices / prices.shift(1)).dropna()

mean_lr = float(log_returns.mean())
sd_lr   = float(log_returns.std())

print("S&P 500 Daily Log-Returns")
print(f"Sample period: {start_date} to {end_date}")
print(f"Mean: {mean_lr}")
print(f"Standard deviation: {sd_lr}")
```

```python
# Download historical data for the tech stocks and save to CSV

yf.Tickers(tech_tickers).history(period='10y', interval='1d')['Close'].to_csv('C:/actuarial_projects/projects/financial_mathematics/normal_vs_mixture/stock_data.csv')
```
